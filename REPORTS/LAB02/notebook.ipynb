# Import thư viện

```python
import pandas as pd
import scipy.stats as stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt
```

# Nhập dữ liệu

```python
group1 = [6.5,6.8,6.3,7.1,7.8,7.5,7.9]
group2 = [5.2,5.9,5,6.1,7.1]
group3 = [6.3,6.4,6.7,6.9,6.1,7.1]
```

# One-way ANOVA

```python
stats.f_oneway(group1, group2, group3)
```

# Chuẩn bị dữ liệu cho Tukey

```python
data = group1 + group2 + group3

groups = (["Nhom1"]*7 +
          ["Nhom2"]*5 +
          ["Nhom3"]*6)
```

# Tukey HSD

```python
tukey = pairwise_tukeyhsd(data, groups, alpha=0.05)

print(tukey)
```

# Vẽ boxplot minh họa

```python
plt.boxplot([group1,group2,group3],
            labels=["Nhom1","Nhom2","Nhom3"])

plt.title("Boxplot of Groups")
plt.xlabel("Group")
plt.ylabel("Value")

plt.show()

plt.savefig("python_boxplot.png")
```
